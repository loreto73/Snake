```python
import turtle
import time
import random

# Configuración de la pantalla
screen = turtle.Screen()
screen.title("Juego de la Serpiente")
screen.setup(600, 600)
screen.bgcolor("black")
screen.tracer(0)

# Configuración de la serpiente
cabeza = turtle.Turtle()
cabeza.shape("square")
cabeza.color("white")
cabeza.penup()
cabeza.goto(0, 0)
cabeza.direction = "stop"

# Configuración de la comida
comida = turtle.Turtle()
comida.shape("circle")
comida.color("red")
comida.penup()

def generar_comida():
    """
    Genera una nueva posición aleatoria para la comida, asegurándose de que no se genere dentro del cuerpo de la serpiente.

    Esta función se encarga de calcular coordenadas aleatorias para la comida y verifica que no coincidan con la posición de ningún segmento de la serpiente. Si la posición generada es inválida, se genera una nueva posición hasta encontrar una válida.
    """

    # Generamos coordenadas aleatorias dentro de los límites de la pantalla
    x = random.randint(-280, 280)
    y = random.randint(-280, 280)

    # Verificamos que la posición generada no coincida con ningún segmento de la serpiente
    while (x, y) in [(segmento.xcor(), segmento.ycor()) for segmento in segmentos]:
        x = random.randint(-280, 280)
        y = random.randint(-280, 280)

    # Movemos la comida a la nueva posición
    comida.goto(x, y)

segmentos = []  # Lista para almacenar los segmentos del cuerpo

# Funciones
def arriba():
    if cabeza.direction != "down":
        cabeza.direction = "up"

def abajo():
    if cabeza.direction != "up":
        cabeza.direction = "down"

def izquierda():
    if cabeza.direction != "right":
        cabeza.direction = "left"

def derecha():
    if cabeza.direction != "left":
        cabeza.direction = "right"

def mover():
    """
    Actualiza las posiciones de la cabeza y los segmentos del cuerpo de la serpiente.

    Esta función calcula las nuevas coordenadas de la cabeza de la serpiente en función de su dirección actual. 
    Luego, actualiza las posiciones de los segmentos del cuerpo para que sigan a la cabeza.
    """

    # Actualizamos la posición de la cabeza
    if cabeza.direction == "up":
        y = cabeza.ycor()
        cabeza.sety(y + 20)
    elif cabeza.direction == "down":
        y = cabeza.ycor()
        cabeza.sety(y - 20)
    elif cabeza.direction == "left":
        x = cabeza.xcor()
        cabeza.setx(x - 20)
    elif cabeza.direction == "right":
        x = cabeza.xcor()
        cabeza.setx(x + 20)

    # Actualizamos la posición de los segmentos del cuerpo
    for index in range(len(segmentos) - 1, 0, -1):
        x = segmentos[index - 1].xcor()
        y = segmentos[index - 1].ycor()
        segmentos[index].goto(x, y)

    # El primer segmento sigue a la cabeza
    if segmentos:
        x = cabeza.xcor()
        y = cabeza.ycor()
        segmentos[0].goto(x, y)

def crecer():
    # Agregamos un nuevo segmento al inicio de la lista
    nuevo_segmento = turtle.Turtle()
    nuevo_segmento.shape("square")
    nuevo_segmento.color("grey")
    nuevo_segmento.penup()
    segmentos.insert(0, nuevo_segmento)

def generar_comida():
    """
    Genera una nueva posición aleatoria para la comida, asegurándose de que no se genere dentro del cuerpo de la serpiente.

    Esta función se encarga de calcular coordenadas aleatorias para la comida y verifica que no coincidan con la posición de ningún segmento de la serpiente. Si la posición generada es inválida, se genera una nueva posición hasta encontrar una válida.
    """

    # Generamos coordenadas aleatorias dentro de los límites de la pantalla
    x = random.randint(-280, 280)
    y = random.randint(-280, 280)

    # Verificamos que la posición generada no coincida con ningún segmento de la serpiente
    while (x, y) in [(segmento.xcor(), segmento.ycor()) for segmento in segmentos]:
        x = random.randint(-280, 280)
        y = random.randint(-280, 280)

    # Movemos la comida a la nueva posición
    comida.goto(x, y)

# Control del teclado
screen.listen()
screen.onkeypress(arriba, "Up")
screen.onkeypress(abajo, "Down")
screen.onkeypress(izquierda, "Left")
screen.onkeypress(derecha, "Right")

# Bucle principal del juego
while True:
    screen.update()

    # Colisión con la comida
    if cabeza.distance(comida) < 20:
        generar_comida()
        crecer()

    # Mover la serpiente
    mover()

    # Comprobar colisiones con los bordes y con el cuerpo
    if cabeza.xcor() > 290 or cabeza.xcor() < -290 or cabeza.ycor() > 290 or cabeza.ycor() < -290:
        # Fin del juego
        print("¡Has chocado con la pared! Fin del juego.")
        time.sleep(1)
        screen.bye()

    for segmento in segmentos[1:]:  # Comenzamos desde el segundo segmento para evitar colisiones inmediatas
        if cabeza.distance(segmento) < 10:
            # Fin del juego
            print("¡Has chocado contigo mismo! Fin del juego.")
            time.sleep(1)
            screen.bye()

    # Aumentar gradualmente la velocidad
    time.sleep(0.1 - 0.001 * len(segmentos))
```
